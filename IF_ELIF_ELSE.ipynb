#**Percabangan/Struktur Pemilihan**

Pada bagian ini, kamu akan membuat keputusan pada bahasa pemrograman Phyton menggunakan struktur percabangan atau juga sering disebut IF ELSE Statment. Pengambilan keputusan dibutuhkan ketika kamu akan mengeksekusi suatu kode jika sebuah kondisi tertentu dipenuhi.

Statment ```if .. elif ... else``` digunakan pada bahasa Phyton untuk pengambilan keputusan.

#**IF Statment**

**Phyton IF Statment Syntax**

````python
if test expression:
  statment(s)
````

**Example Phyton IF Statment**

````python
#Jika angka adalah bilangan posiif maka print "adalah bilangan positif"
Angka = 3
if Angka > 0:
  print(Angka, "adalah bilangan posistif")
print("pernyataan ini selalu ditampilkan")

Angka = -1
if Angka > 0:
  print(Angka, "adalah bilangan posistif")
print("pernyataan ini juga selalu ditampilkan")
````

#**IF ELSE Statments**

**Phyton IF ELSE Statment Syntax**

````python
if test expression:
  Body of if
else:
  Body of else
````

**Phyton IF ELSE Statment Flowchart**

Screenshot_20230705_234021_Word.jpg

**Example Phyton IF ELSE Statment**

````python
#Program disusun untuk mengecek apakah angka tersebut kurang dari 0 atau tidak
#Dan menampilkan tulisan yang sesuai

Angka = 3

if Angka >= 0:
  print("Bilangan lebih dari atau samadengan 0")
else:
  print("Bilangan kurang dari 0")
````

````python
#Program disusun untuk mengecek apakah dua buah angka sama atau tidak
#Dan menampilkan tulisan yang sesuai

a = 5
b = 5

if a==b:
  print("a dan b bernilai sama")
else:
  print("a dan b bernilai berbeda")
  print("belajar if else")
````
