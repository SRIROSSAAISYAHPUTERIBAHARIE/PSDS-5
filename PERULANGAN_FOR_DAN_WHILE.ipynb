#**Looping/Perulangan**

Phyton menyediakan dua statment untuk melakukan proses perulangan yaitu for dan while. Diantara kedua statment ini, secara umum for lebih banyak digunakan daripada while.

#**For Loop**

Perulangan ```for``` digunakan untuk melakukan iterasi di dalam ```sequence``` seperti ```list, tuple, string```, dan ```array``` atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**

````python
for val in sequence:
  loop body
````

**For Loop Flochart**

Screenshot_20230705_190236_Word.jpg

**Example Phyton For Loop**

````python
#Program untuk menghitung jumlah semua angka dalam sebuah list
L1 = [2, 4, 5, 7, 9]
Jumlah = 0

#Iterasi di dalam list
for i in L1:
  Jumlah = Jumlah+i

print(Jumlah)
````

#**The Range() Function**

````python
#Program untuk melakukan iterasi di dalam list melalui indeks
Minuman = ["Teh", "Boba", "Kopi", "Susu"]
P = len(Minuman)

#Iterasi di dalam list melalui indeks
for i in range(P):
  print(Minuman[i])
````

#**For Loop With Else**

````python
#Contoh for Loop menggunakan else
for i in range(P):
  print(Minuman[i])
else:
  print("Elemen sudah habis")
````

#**Latihan 1**

Diberikan ```List_Tagihan```. Kamu diminta untuk menghitung jumlah tagihan!. Gunakan Looping ```for``` untuk menyelesaikan tugas ini!
Simpan hasilnya pada variabel ```Total_Tagihan```!

```List_Tagihan = [50000, 75000, -150000, 125000, 30000, -50000, 200000]```

````python
#Program untuk menghitung jumlah semua angka dalam sebuah list
List_Tagihan = [50000, 75000, -150000, 125000, 30000, -50000, 200000]
Total_Tagihan = 0

#Iterasi di dalam list
for i in List_Tagihan:
  Total_Tagihan = Total_Tagihan+i

print(Total_Tagihan)
````

#**While Loop**

While Loop pada Phyton digunakan untuk melakukan iterasi pada blok kode selama nilai dari ```Test Expresion``` adalah ```True```.

**Syntax of While Loop in Phyton**

````python
while Test_Expresion:
  Body of while
````

**While Loop Flowchart**

Screenshot_20230705_190250_Word.jpg

**Example Phyton While Loop**

````python
#Program untuk menjumlahkan semua bilangan asli hingga suku ke-n
#Jum = 1+2+3+...+n

n = 10
Jumlah = 0
i = 1

while i <= n:
  Jumlah = Jumlah+i
  i = i+1
print(Jumlah)
````

#**While Loop with Else**

Sama dengan ```for``` loop kita juga dapat mengkombinasikan ```while``` loop dengan blok ```else```.

Bagian ```else``` akan dieksekusi jika nilai pada ```Test_Expresion``` pada ```while loop``` bernilai ```false```. Perhatikan contoh berikut:

````python
Jum = 0
i = 1
while i <= n:
  Jum = Jum+i
  i = i+1
else:
  print(Jum) #Dieksekusi diluar looping
````

#**Latihan 2**

Diberikan list ```Tagihan``` hitunglah ```Total_Tagihan``` menggunakan ```While Loop```.

```Tagihan = [50000, 75000, 125000, 300000, 200000]```

````python
Tagihan = [50000, 75000, 125000, 300000, 200000]
#Inisialisasi variabel total_tagihan
Total_Tagihan = 0

#Menggunakan while loop untuk menghitung total tagihan
indeks = 0
while indeks < len(Tagihan):
  Total_Tagihan += Tagihan[indeks]
  indeks += 1

#Menampilkan total tagihan
print("Total_Tagihan : ", Total_Tagihan)
````
